```python
import pandas as pd
import numpy as np
```

```python
df_test = pd.read_csv('/content/drive/MyDrive/Data/test_data_2.csv')
```

```python
model = ClassificationModel("bert", "/content/drive/MyDrive/Models/BERT_large")
```

```python
labels = np.full((len(df_test)), -1)
for i in range(83):
  n = 1000 # labels data in batches of 1000 tweets
  print(i)
  test_text=list(df_test['text'][n*i:n*(i+1)])
  predictions, raw_outputs = model.predict(test_text)
  labels[n*i:n*(i+1)] = predictions
# label the remaining tweets
test_text=list(df_test['text'][n*(i+1):])
predictions, raw_outputs = model.predict(test_text)
labels[n*(i+1):] = predictions
# save labels
np.save('/content/drive/MyDrive/Data/labels/labels.npy', labels)
```
